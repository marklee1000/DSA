```python
import statistics

def choose_order(menu_nav):
    order_choice = int(input("\nWhat type of arrangement?\n[1] Descending Order\n[2] Ascending Order\n[3] Back\nEnter your choice: "))
    if order_choice == 1:
        return 1
    elif order_choice == 2:
        return 2
    elif order_choice == 3:
        menu_bar(nlist)

def menu_bar(nlist, menu_loop):
    menu_nav = int(input("\nMENU:\n[1] Bubble Sort\n[2] Selection Sort\n[3] Insertion Sort\n[4] Exit\nEnter your choice: "))

    if menu_nav == 1:
        arrangement = choose_order(menu_nav)
        bubble_sort(nlist, arrangement)
    elif menu_nav == 2:
        arrangement = choose_order(menu_nav)
        select_sort(nlist, arrangement, len(nlist))
    elif menu_nav == 3:
        arrangement = choose_order(menu_nav)
        insert_sort(nlist, arrangement)
    elif menu_nav == 4:
        menu_loop == 0

def stats(nlist, size, swaps):
    mean = sum(nlist) / size
    median = statistics.median(nlist)
    print(f"\nMean: {mean}")
    print(f"Median: {median}")
    print(f"Number of swaps: {swaps}")

def bubble_sort(nlist, arrangement):
    swaps = 0
    size = len(nlist)
    for i in range(size):
        for j in range(0, size - i - 1):
            if arrangement == 1:
                if nlist[j] < nlist[j + 1]:
                    temp = nlist[j]
                    nlist[j] = nlist[j + 1]
                    nlist[j + 1] = temp
                    print(nlist)
                    swaps += 1
            elif arrangement == 2:
                if nlist[j] > nlist[j + 1]:
                    temp = nlist[j]
                    nlist[j] = nlist[j + 1]
                    nlist[j + 1] = temp
                    print(nlist)
                    swaps += 1
    print(f"\nFinal List: {nlist}")
    stats(nlist, size, swaps)

def select_sort(nlist, arrangement, size):
    swaps = 0
    for step in range(size):
        min_idx = step
        for i in range(step + 1, size):
            if arrangement == 1:
                if nlist[i] > nlist[min_idx]:
                    min_idx = i
                    print(nlist)
                    swaps += 1
            elif arrangement == 2:
                if nlist[i] < nlist[min_idx]:
                    min_idx = i
                    print(nlist)
                    swaps += 1
        (nlist[step], nlist[min_idx]) = (nlist[min_idx], nlist[step])
    print(f"\nFinal List: {nlist}")
    stats(nlist, size, swaps)

def insert_sort(nlist, arrangement):
    swaps = 0
    size = len(nlist)
    for step in range(1, size):
        key = nlist[step]
        j = step - 1
        if arrangement == 1:
            while j >= 0 and key > nlist[j]:
                nlist[j + 1] = nlist[j]
                j = j - 1
                print(nlist)
                swaps += 1
        elif arrangement == 2:
            while j >= 0 and key < nlist[j]:
                nlist[j + 1] = nlist[j]
                j = j - 1
                print(nlist)
                swaps += 1
        nlist[j + 1] = key
    print(f"\nFinal List: {nlist}")
    stats(nlist, size, swaps)

def main():
    nlist = []
    while True:
        num = int(input("Enter a number (enter 0 to stop): "))
        if num == 0:
            break
        nlist.append(num)
    print(nlist)
    size = len(nlist)
    menu_loop = 1
    while menu_loop == 1:
        menu_bar(nlist, menu_loop)
        menu_check = input("\nDo you want to try again? [Y/N] ")
        if menu_check.lower() == 'y':
            main()
        elif menu_check.lower() == 'n':
            print("\nThank you for using the program!")
            menu_loop = 0

main()
```
