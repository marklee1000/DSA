C = 2*TB + 5A; 
D = (TC + B) - 7TA; 
E = 3*(8 + TD)**(2) + TA - 7; 

```python
print("A:")
A=[]
for i in range(3):
    a=[]
    for j in range(3):
        a.append(int(input()))
    A.append(a)
for i in range (3):
    for j in range (3):
        print(A[i][j], end=" ")
    print()

print("\nB:")
B=[]
for i in range(3):
    a=[]
    for j in range(3):
        a.append(int(input()))
    B.append(a)
for i in range (3):
    for j in range (3):
        print(B[i][j], end=" ")
    print()

print("\nTransposed A:")
TA = []
for i in range (3):
    a=[]
    for j in range (3):
        t=A[j][i]
        a.append(t)
    TA.append(a)
for i in range(3):
    for j in range(3):
        print(TA[i][j], end=" ")
    print()

print("\nTransposed B:")
TB = []
for i in range (3):
    a=[]
    for j in range (3):
        t=B[j][i]
        a.append(t)
    TB.append(a)
for i in range(3):
    for j in range(3):
        print(TB[i][j], end=" ")
    print()

print("\nC:")
C=[]
for i in range (3):
    a=[]
    for j in range (3):
        c=2*TB[i][j]+5*A[i][j]
        a.append(c)
    C.append(a)
for i in range (3):
    for j in range (3):
        print(C[i][j], end = " ")
    print()
print()

print("\nTransposed C:")
TC = []
for i in range (3):
    a=[]
    for j in range (3):
        t=C[j][i]
        a.append(t)
    TC.append(a)
for i in range(3):
    for j in range(3):
        print(TC[i][j], end=" ")
    print()
    
print("\nD:")
D=[]
for i in range (3):
    a=[]
    for j in range (3):
        d=(TC[i][j]+B[i][j])-7*TA[i][j]
        a.append(d)
    D.append(a)
for i in range (3):
    for j in range (3):
        print(D[i][j], end = " ")
    print()
print()

print("\nTransposed D:")
TD = []
for i in range (3):
    a=[]
    for j in range (3):
        t=D[j][i]
        a.append(t)
    TD.append(a)
for i in range(3):
    for j in range(3):
        print(TD[i][j], end=" ")
    print()

print("\nE:")
E=[]
for i in range (3):
    a=[]
    for j in range (3):
        e=3*(8 + TD[i][j])**(2) + TA[i][j] - 7
        a.append(e)
    E.append(a)
for i in range (3):
    for j in range (3):
        print(E[i][j], end = " ")
    print()
print()
```
